* Now we import KMeans from sklearn.cluster module and to find the value of k we use silhouette score or elbow method
* From silhouette score we get the exact value and the verification of k value.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df
```

* To get which k value is best we pass a loop and find the inertia of all the datapoints, we generally select the datapoints whose inertia is low, but here we are also finding the silhouette score, so as per it, the k value who has the highest silhouette score is selected.
* Also we measure the distance between datapoints using euclidean distance.

```python
see=[]   
silhouette={}   
k_range=range(2,15)   
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(x)
    see.append(km.inertia_)
    silhouette[k]=silhouette_score(x,km.labels_,metric='euclidean')
```

```python
km.labels_
```

```python
km.inertia_
```

```python
silhouette
```

* We are plotting a line graph of k values v/s WCSS(within column square os sums which is nothing but inertia)

```python
plt.plot(k_range,see,color='red',marker='*')
plt.xlabel('values of k')
plt.ylabel('WCSS')
plt.show()
```

* As we can see that k value of 2 give a good clusters, passing it into the KMeans n cluster value.

```python
km=KMeans(n_clusters=2) 
ypred=km.fit_predict(x)
ypred
```

```python
df['Cluster'] = ypred
```

```python
x
```

```python
df1=df[df['Cluster']==0]
df2=df[df['Cluster']==1]
```

```python
km.cluster_centers_
```

```python
Years=df[['2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']]
plt.figure(figsize=(12,9))
plt.scatter(df1['Country'],df1['Inflation Rate'],color='green')
plt.scatter(df2['Country'],df2['Inflation Rate'],color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='red',marker='x')
plt.legend()
```

```python
cmap = {0 : 'grey',1 : 'green'}
label_color = [cmap[l] for l in km.labels_]
plt.figure(figsize=(12,9))
plt.scatter([i[0] for i in x], [i[1] for i in x], c=label_color)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='black', marker='D')
plt.show()
```
